```python
import numpy
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
%matplotlib inline
```

```python
low_df=pd.read_csv('/home/will/Projects/language_laterality/10/language.csv')
str_low_df=low_df.loc[low_df['Zeros']!=66]
high_df=pd.read_csv('/home/will/Projects/language_laterality/05/master.csv')
str_high_df=high_df.loc[high_df['Zeros']!=66]
highest_df=pd.read_csv('/home/will/Projects/language_laterality/01/language.csv')
str_highest_df=highest_df.loc[highest_df['Zeros']!=66]
df_list=[low_df, high_df, highest_df]
whole_df=pd.concat(df_list, ignore_index=True)
whole_df = whole_df.loc[whole_df['Zeros']!=66]

low_sm_df=pd.read_csv('/home/will/Projects/language_laterality/10/scenemem.csv')
str_low_sm_df=low_sm_df.loc[low_sm_df['Zeros']!=30]
high_sm_df=pd.read_csv('/home/will/Projects/language_laterality/05/scenemem.csv')
str_high_sm_df=high_sm_df.loc[high_sm_df['Zeros']!=30]
highest_sm_df=pd.read_csv('/home/will/Projects/language_laterality/01/scenemem.csv')
str_highest_sm_df=highest_sm_df.loc[highest_sm_df['Zeros']!=30]
df_sm_list=[low_sm_df,high_sm_df, highest_sm_df]
whole_sm_df = pd.concat(df_sm_list, ignore_index=True)
whole_sm_df = whole_sm_df.loc[whole_sm_df['Zeros']!=30]


```

```python
test = [str_low_sm_df, str_high_sm_df]
test_df = pd.concat(test)
pivoted=test_df.pivot(columns='Alpha', values='hippocampus LI')

test2 = [str_low_sm_df, str_highest_sm_df]
test_df2 = pd.concat(test2)
pivoted2=test_df2.pivot(columns='Alpha', values='hippocampus LI')

test3 = [str_high_sm_df, str_highest_sm_df]
test_df3 = pd.concat(test3)
pivoted3=test_df3.pivot(columns='Alpha', values='hippocampus LI')
```

```python
def comparison_plot(roi, xlim):
    kwargs = dict(hist_kws={'alpha':.35}, kde_kws={'linewidth':1.5})
    sns.distplot( low_df[roi],  color="green", label="0.10", **kwargs)
    sns.distplot( high_df[roi],  color="blue", label="0.05", **kwargs)
    sns.distplot( highest_df[roi],  color="red", label="0.01", **kwargs)
    plt.legend()
    plt.xlim(xlim)
    plt.show()
```

```python
def str_comparison_plot(roi, xlim):
    kwargs = dict(hist_kws={'alpha':.35}, kde_kws={'linewidth':1.5})
    sns.distplot( str_low_df[roi],  color="green", label="0.10", **kwargs)
    sns.distplot( str_high_df[roi],  color="blue", label="0.05", **kwargs)
    sns.distplot( str_highest_df[roi],  color="red", label="0.01", **kwargs)
    plt.legend()
    plt.xlim(xlim)
    plt.show()
```

```python
comparison_plot("broca's area LI", [-1,1])
```
